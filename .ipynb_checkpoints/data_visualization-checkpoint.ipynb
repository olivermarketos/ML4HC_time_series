```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data_test = pd.read_parquet('./data.parquet/test')
data_validate = pd.read_parquet('./data.parquet/validate')
data_test_and_validate = pd.concat([data_test, data_validate], ignore_index=True)
```

```python
data_test
```

```python
# General plots of the variables
plt.figure(figsize=(24, 16))

for i, column in enumerate(data_test.columns.drop(['RecordID', 'Time', 'MechVent', 'ICUType'])):
    column_values = data_test_and_validate[column].dropna().to_numpy()
    plt.subplot(8, 5, i+1)
    sns.histplot(column_values, bins=30)
    plt.title(column)

plt.tight_layout()
plt.savefig('./visuals/distributions.png')
plt.show()
```

```python
# Plots based on gender
plt.figure(figsize=(24, 16))

for i, column in enumerate(data_test.columns.drop(['RecordID', 'Time', 'MechVent', 'ICUType', 'Gender'])):
    plt.subplot(8, 5, i+1)
    sns.histplot(data=data_test_and_validate, x=column, hue='Gender', alpha =.5, bins=30)
    plt.title(column)


plt.tight_layout()
plt.savefig('./visuals/gender_distribution.png')
plt.show()
```

```python
# Plots based on age groups
data_test_and_validate['age_group'] = pd.cut(data_test_and_validate['Age'], bins=range(15, 95, 10), right=True)
plt.figure(figsize=(24, 16))

for i, column in enumerate(data_test.columns.drop(['RecordID', 'Time', 'MechVent', 'ICUType', 'Age'])):
    plt.subplot(8, 5, i+1)
    sns.histplot(data=data_test_and_validate, x=column, hue='age_group', alpha =.5, bins=30)
    plt.title(column)


plt.tight_layout()
plt.savefig('./visuals/age_distribution.png')
plt.show()
```

```python
# Plots based on ICU type
plt.figure(figsize=(24, 16))

for i, column in enumerate(data_test.columns.drop(['RecordID', 'Time', 'MechVent', 'ICUType'])):
    plt.subplot(8, 5, i+1)
    sns.histplot(data=data_test_and_validate, x=column, hue='ICUType', alpha =.5, bins=30)
    plt.title(column)


plt.tight_layout()
plt.savefig('./visuals/icu_distribution.png')
plt.show()
```
